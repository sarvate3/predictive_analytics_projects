```python
# return total combined value of items
# given set of items' values and count vector
def calculate_item_value(values, item_count_vector):
    total_val = 0
    for val, units in zip(values, item_count_vector):
        total_val += val * units
    return(total_val)
```

```python
def Alg1(weights, values, capacity, item_count_vector):
    print(item_count_vector)
    # choose items while there is remaining capacity 
    # for at least one item
    number_items = len(weights)
    max_value_index = -1
    if (min(weights) <= capacity):
            max_value = 0
            # loop through each item and select one with
            # maximum value, provided not exceeding capacity
            for i in range(1, number_items):
                if (weights[i] <= capacity):
                    if(values[i] > max_value):
                        max_value_index = i
                        max_value = values[max_value_index]
                else:
                    continue
            # increase item count by one for max value item determined in above loop
            # reduce capacity accordingly based on item weight
            # max_value_index = -1 is indicator that no compatible item was found
            if (max_value_index != -1):
                item_count_vector[max_value_index] += 1
                capacity -= weights[max_value_index]
    # recursively call function wth new item count vector and capacity
    # if there is remaining capacity for at least one item
    if (min(weights) <= capacity):
        Alg1(weights, values, capacity, item_count_vector)
    return(item_count_vector)
```

```python
def Alg2(weights, values, capacity, item_count_vector):
    # initialize empty item state vector
    item_state = []
    number_items = len(weights)
    #for each item, start with max allocation to item, given total capacity
    for i in range(0, number_items):
            item_count_vector[i] = math.floor(capacity/weights[i])
            
            #item_state_ijk = [item_count_vector, calculate_item_value(values, item_count_vector)]
            #print(item_state_ijk)
            
            start_index = item_count_vector[i]
            
            # decrement count of item i all the way to 1
            for j in range(0, start_index+1):
                
                # for each unit decrement, reallocate capacity to every other item
                # one by one based on amount of weight needed to be reallocated
                if j > 0:
                    item_count_vector[i] -= 1
                
                
                if (j == 0):
                    extra_weight = 0
                else:
                    extra_weight += weights[i]
                
                print(" ")
                print("item {} decrement: {}".format(i, j))
                print("item {} count: {}".format(i, item_count_vector[i]))
                print("Extra weight available: {}".format(extra_weight))
                
                

                for k in range(0, number_items):
                    max_value = 0
                    # skip for k = i, i.e. same alternate item as one being decremented
                    if (k==i):
                        continue
                    else:
                        # increase item k count by extra_weight
                        item_k_increment = math.floor(extra_weight/weights[k])
                        item_count_vector[k] += item_k_increment
                        
                        print(" ")
                        print("item {} increment: {}".format(k, item_k_increment))
                        print("item {} count: {}".format(k, item_count_vector[k]))
                        print("Extra weight allocated: {}".format(weights[k] * item_k_increment))
                        print("Extra value added: {}".format(values[k] * item_k_increment))
                        
                        #item_state_ijk = [item_count_vector, calculate_item_value(values, item_count_vector)]
                        #print(item_state_ijk)
                        
                        
                        # check total value of items
                        # if greater than last max value, update index
                        if (calculate_item_value(values, item_count_vector) > max_value):
                            max_value_count_k = item_k_increment
                        # reverse item k increment and repeat 
                        # until end of loop which increments every item k
                        item_count_vector[k] -= item_k_increment
                        

                        #print(item_count_vector[k])
                
                # after all k item increments tested
                # record item state vector which produced maximum value
                # reverse item j decrement and repeat
                # until end of loop which decrements item i to 0
                
                
                item_count_vector[k] += max_value_count_k
                print(" ")
                print("Incremented item {} by {}".format(k, max_value_count_k))
                print("New count of item {}: {}".format(k, item_count_vector[k]))
                
                # decrease extra weight variable by amount equal to weight increase
                # by incrementing k
                        
                extra_weight -= item_k_increment * weights[k]
                        
                
                item_state_ijk = [item_count_vector, calculate_item_value(values, item_count_vector)]
                print(item_state_ijk)
                
                #item_state.append(item_state_ijk)
  
            # after item i tested (from full allocation to zero)
            # reverse item i, k count to zero and test next item at full allocation
    
            item_count_vector[i] = 0
            item_count_vector[k] = 0
            

    # after all tests completed, return item state vector
    return(item_state)
```
